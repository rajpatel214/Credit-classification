```python
def correlation_plot(df):
    
    corr = abs(df.corr()) # correlation matrix
    lower_triangle = np.tril(corr, k = -1)  # select only the lower triangle of the correlation matrix
    mask = lower_triangle == 0  # to mask the upper triangle in the following heatmap

    plt.figure(figsize = (15,10))  # setting the figure size
    sns.set_style(style = 'white')  # Setting it to white so that we do not see the grid lines 
    sns.heatmap(lower_triangle, center=0.5, cmap= 'Blues', xticklabels = corr.index,
                yticklabels = corr.columns,cbar = False, annot= True, linewidths= 1, mask = mask)   # Da Heatmap
    plt.show()

correlation_plot(new_gc)
```

```python
new_columns=gc[['checkin_acc','credit_history','savings_acc','present_emp_since','inst_plans']]

from sklearn.preprocessing import OneHotEncoder
```

```python
ohe= OneHotEncoder(drop='first',sparse_output=False)
```

```python
new_1=ohe.fit_transform(gc[['checkin_acc','credit_history','savings_acc','present_emp_since','inst_plans']])
```

```python
new_gc=pd.DataFrame(new_1,columns=ohe.get_feature_names_out(new_columns.columns))
```

```python
gc.drop(['checkin_acc','credit_history','savings_acc','present_emp_since','inst_plans'],axis=1,inplace=True)
```

```python
gc1=pd.concat([gc,new_gc],axis=1)
```

```python
gc1
```

```python
x=gc1.drop('status',axis=1)

y=gc1['status']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8)
```

```python
from sklearn.preprocessing import StandardScaler

ss=StandardScaler()

sacling_train_gc=ss.fit_transform(x_train)

sacling_test_gc=ss.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score,KFold
```

```python
k=KFold(n_splits=4)

list = [LogisticRegression(),SVC(),DecisionTreeClassifier(),RandomForestClassifier(),KNeighborsClassifier()]

for i in list:
    print(i)
    print(cross_val_score(i,x_train,y_train,cv=k).mean())
```

```python
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression()

lr.fit(x_train,y_train)

lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

 # Metrics

# VotingClassifier

```python
from sklearn.ensemble import VotingClassifier,StackingClassifier,BaggingClassifier
```

```python
log_clf = LogisticRegression()
knn_clf = KNeighborsClassifier()
rendom_clf= RandomForestClassifier(n_estimators=500,max_depth=200,min_samples_split=250)
```

```python
vc=VotingClassifier( estimators=[
        ('lr', log_clf),
        ('knn', knn_clf),
    ('random',rendom_clf)
    ],
    voting='hard')
```

```python
vc.fit(x_train,y_train)
```

```python
vc.score(x_train,y_train)
```

```python
vc.score(x_test,y_test)
```
